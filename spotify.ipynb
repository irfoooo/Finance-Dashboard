```python
most_popular=tracks.query('popularity>90',inplace =False).sort_values("popularity",ascending =False)
most_popular[:10]
```

```python
tracks = tracks.rename(columns={"c": "Item", "customer_type": "Membership"})
df.head()
```

```python
tracks.set_index('release_date', inplace =True)
tracks.index=pd.to_datetime(tracks.index)
tracks.head()
```

```python
tracks["duration"]= tracks["duration_ms"].apply(lambda x: round(x/1000))
tracks.drop("duration_ms",inplace=True, axis=1)
```

```python
tracks.duration.head()
```

```python
corr_df=tracks.drop(["key","mode","explicit"],axis=1).corr(method="pearson")
plt.figure(figsize=(14,6))
heatmap=sns.heatmap(corr_df,annot=True,fmt=".1g",vmin=-1,vmax=1,center=0,cmap="Greens",linewidths=1,linecolor="black" )
heatmap.set_title("Correlation Heatmap Between Variable")
heatmap.set_xticklabels(heatmap.get_xticklabels(),rotation=90)
```

```python
sample_df=tracks.sample(int(0.004*len(tracks)))
```

```python
print(len(sample_df))
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data = sample_df, y = "loudness",x = "energy", color = "g").set(title ="Loudness vs Energy")
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data =sample_df,y= "popularity", x= "acousticness", color ="g").set(title ="Popularity vs Acousticness")
```

```python
tracks['dates']=tracks.index.get_level_values('release_date')
tracks.dates=pd.to_datetime(tracks.dates)
years=tracks.dates.dt.year
```

```python
sns.displot(years, discrete=True,aspect=2,height=5 , kind="hist" ,color='green').set(title="Number of songs per year")
```

```python

total_dr = tracks.duration
fig_dims = (18,7)
fig , ax = plt.subplots(figsize = fig_dims)
fig = sns.barplot(x = years, y = total_dr, ax = ax , errwidth = False ).set(title="Year vs Duration")
plt.xticks(rotation=90)
plt.show
```

```python
df_genre=pd.read_csv("/content/spotifyfeatures1.csv")
df_genre.head()
```

```python
plt.title("Duraton of the Songs in Different Genres")
sns.color_palette("rocket",as_cmap=True)
sns.barplot(y='genre',x='duration_ms',data=df_genre)
plt.xlabel("Duration in milli seconds")
plt.ylabel("Genres")
```

```python
sns.set_style(style = "darkgrid")
plt.figure(figsize=(10,5))
famous = df_genre.sort_values("popularity",ascending = False).head(10)
sns.barplot(y='genre',x='popularity', data = famous).set(title= "Top 5 Genre by Popularity")
```
